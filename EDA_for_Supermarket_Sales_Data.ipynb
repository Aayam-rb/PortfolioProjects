### Univariate Analysis

```python
# finding the aggregate sales among the branches

sns.countplot(df,x='Branch', stat='count', hue='Branch',palette='Set2')    
plt.title('Numbers of Sales Transaction by Branch')
plt.show()

print('Branch A has the highest number of transactions (340), followed closely by Branch B (332) and Branch C (328),')
print("indicating similar performance across all branches with only minor differences in transaction counts.")
print(df['Branch'].value_counts())
```

```python
# Find the most popular payment method used by the customers

plt.figure(figsize=(7, 7))
df['Payment'].value_counts().plot(kind='pie',
                                  autopct='%1.2f%%',
                                  startangle=90,
                                  colors=['#FF6666', '#66FF66', '#66B3FF'],  # Custom colors
                                       wedgeprops={'edgecolor': 'grey', 'linewidth': 1}, # Adding black border around slices
                                       textprops={'fontsize': 12, 'fontweight': 'bold'})  # Enhance text readability)

plt.title('Payment Methods Used by Customers')
plt.show()

print('Ewallet has the highest number of transactions (34.50%), followed by Cash (34.40), and Credit card (31.10),')
print('indicating similar transaction volumes across all payment methods with a slight lead for Ewallet.')
```

```python
#Find the Distribution of customer ratings

ax = sns.distplot(df['Rating'])
ax.set_xlabel('Ratings')
plt.show()
print('Skewness=', df['Rating'].skew())
print('Since the distribution of the ratings are almost zero it indicates a normal distribution. ')
print('Which means the ratings are balanced and there is no sign of any biasness towards higher or lower values')
```

### Bivariate / Multivariate Analysis

```python
# Does the Cost of goods sold afftect the ratings that the customers provide?
# is there any realtion between the Cogs and the Customers Ratings?

ax = df.plot(kind='scatter',x='Rating',y='cogs',title='Relation between Cogs and Customer Ratings')
ax.set_xlabel('Customer Ratings')
ax.set_ylabel('Cost of Goods Sold')
plt.show()
print('Since there is no clear patterns or trends in the scatterplot, it suggests that there is no significant relation between Cost of Goods Sold and Customer Ratings ')
```

```python
# Does the Branch of the supermarket have anything to do with the Customers Ratings?

ax = sns.barplot(x='Branch', y='Rating', data= df,  hue='Branch',palette='Set2' )

plt.show()
print("There does not appear to be a significant relation between Branch and Customer Ratings ")
```

```python
# Analyzing Which branch brings in more gross income

ax = sns.barplot(x='Branch',y='gross income', data=df , palette='Set2')
ax.set_title('Gross income per branch')
ax.set_xlabel('Branch')
ax.set_ylabel('Gross income')
plt.show()
print('As per the Bar plot we can observed that Branch C has the highest Gross income among') 
print('different branches, followed by Branch B and lastly by Branch A.')
print('We can conclude that branch C is the most profitable branch as per gross income.')
```

```python
# Find which city generates the most gross income

ax = sns.barplot(x='Branch',y='gross income', data=df , hue= df['City'],palette='Set2')
ax.set_title('Gross Income by Branch and City')
ax.set_xlabel('Branch')
ax.set_ylabel('Gross income')
plt.show()

print('This shows that the Naypyitaw has the highest gross income.')
```

```python
# Which Gender contributes to the higher gross income

ax = sns.barplot(data=df, x='Gender',y='gross income',hue=df['City'], palette='Set2')
ax.set_title('Gross Income by Gender Across Different Cities')
ax.set_xlabel('Gender')
ax.set_ylabel('Gross Income')
plt.show()
print('According to the above bar chart we can observed that females have more contibutions to the gross income compared to male across the cities')
print('This indicates that gender plays a role n=in gross income generation,with females showing  higher contribution in this dataset.')
```

```python
# Which Product Line generates the most amount of income

sns.barplot(data=df, x='Product line', y='gross income',palette='viridis')
plt.title('Gross income by Product line')
plt.xlabel('Product Line')
plt.ylabel('Gross Income')
plt.xticks(rotation=80)  # Rotate x-axis labels for better readability (if needed)
plt.show()
print('We can observe that the highest amount of gross income is generated by Home and lifestyle  product followed by Sports and travel.')
```

```python
# Which Gender Product Line generates the most amount of income

sns.barplot(data=df, x='Product line', y='gross income',hue=df['Gender'],palette='deep')
plt.title('Gross Income Contribution by Gender Across Product Lines')
plt.xlabel('Product Line')
plt.ylabel('Gross Income')
plt.xticks(rotation=80)  # Rotate x-axis labels for better readability (if needed)
plt.show()

print('Females contribute more to gross income in most product lines, especially in "Home and Lifestyle" and "Food and Beverages," while males contribute more  evenly across all categories.')
```

```python
 # Find the Highest Unit price in the product line

sns.barplot(x='Product line', y='Unit price', data=df,palette='plasma')
plt.title('Unit Price Across Product Lines')
plt.xlabel('Product Line')
plt.ylabel('Unit Price')
plt.xticks(rotation=80)  # Rotate x-axis labels for better readability (if needed)
plt.show()

print('We can observe that the Sports and travel product line has the Highest unit price and followed by other product line  ')
```

```python
# Finding the Different payment Methods used by Customers City wise

print(pd.crosstab(df['City'],df['Payment']))
sns.heatmap(pd.crosstab(df['City'],df['Payment']))
plt.title('Different Payment Methods Used by Customers Across Cities')
plt.show()
#df.groupby(['City','Payment'])['Payment'].count().plot(kind='')

print('The heatmap shows payment methods (Cash, Credit card, Ewallet) across cities. In Mandalay, Cash and Ewallet are similar, with Credit card close to Cash. Naypyitaw has the highest    Cash payments, while Yangon favors Ewallet.The color gradient highlights these differences,  with lighter colors indicating higher values')
```

```python
# Find which Product line is purchased in the highest quantity

print(df.groupby('Product line').sum()['Quantity'])
(df.groupby('Product line').sum()['Quantity']).plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b'])
plt.xticks(rotation=80)
plt.ylabel('Quantity')
plt.show()

print('The data shows the number of products in different product lines, with electronic accessories having the highest count at 971 products.Food and beverages follows closely with 952 products, while fashion accessories has 902 products. Sports and travel comes next with 920 products, followed by home and lifestyle at 911 products. Health and beauty has the lowest count, with 854 products. This suggests that electronic accessories dominate the product range,   while health and beauty products are the least represented.')
```

```python
# daily Sales by day of week
df['Date']=pd.to_datetime(df['Date'])
print(df['Date'].dt.day_name().value_counts())
df['Date'].dt.day_name().value_counts().plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2'])
plt.xticks(rotation=80)
plt.ylabel('Sales')
plt.xlabel('Week Day')
plt.show()

print('The data represents the sales for each day of the week, with the highest sales occurring on Saturday at 164 units, followed by Tuesday at 158 units.     Wednesday and Friday have similar sales figures of 143 and 139, respectively. Thursday comes next with 138 sales, and Sunday follows closely with 133    sales. Monday has the lowest sales at 125 units.This indicates that the sales peak on Saturday, likely due to higher consumer activity towards the end of the week, and gradually decrease as the week progresses, with Monday seeing the least sales.')
```

```python
# Highest Month for Sales

df['Date']=pd.to_datetime(df['Date'])
print(df['Date'].dt.strftime('%B').value_counts()) # %B - represents Full Month name
df['Date'].dt.strftime('%B').value_counts().plot(kind='bar' ,color=['#1f77b4', '#ff7f0e', '#2ca02c',])
plt.xticks(rotation=80)
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()

# Common strftime format codes:
# %Y: Year with century (e.g., 2024)
# %m: Month as a zero-padded decimal number (01-12)
# %d: Day of the month as a zero-padded decimal number (01-31)
# %B: Full month name (e.g., January, February)
# %b: Abbreviated month name (e.g., Jan, Feb)
# %A: Full weekday name (e.g., Monday, Tuesday)
# %a: Abbreviated weekday name (e.g., Mon, Tue)
# %H: Hour in 24-hour format (00-23)
# %I: Hour in 12-hour format (01-12)
# %M: Minute as a zero-padded decimal number (00-59)
# %S: Second as a zero-padded decimal number (00-59)
# %p: AM/PM

print('The sales data indicates that January had the highest sales with 352 units, followed by a dip in February, where sales dropped to 303 units. However,    sales rose again in March to 345 units, suggesting a recovery in sales after the decline in February.')
```
