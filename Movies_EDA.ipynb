```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 20)
```

```python
df = pd.read_csv(r'movies.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

### # Dorpping unecessary columns

```python


df.drop(columns= ['id', 'imdb_id', 'tagline','budget_adj','overview','revenue_adj', 'cast','homepage'] , inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

### # Dropping rows with null values 

```python


df.dropna(how='any' ,subset=['genres','director'], inplace =True)
```

### # Filling null values with 0

```python


df['keywords'] = df['keywords'].fillna(0)
df['production_companies'] = df['production_companies'].fillna(0)
```

```python
df.isnull().sum()
```

### # Rounding off the popularity into 2 decimal spaces

```python


df['popularity'] = df['popularity'].round(2)
```

```python
df
```

### # Adding a new column for ROI

```python


df.insert(3, "profit", df.revenue-df.budget)
```

```python
df.insert(4, "ROI", df.profit/df.budget)
```

```python
df['ROI'] = df.ROI.round(2)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['ROI'].value_counts()
```

### # Since the ROI cloumns contains infinite values we must convert it into NaN values , helps for easier analysis 

```python
non_finite_values = ~np.isfinite(df['ROI'])
```

```python
non_finite_values.sum()

#There are 5636 non finite values , we convert them into NaN 
```

```python
df['ROI'] = df['ROI'].replace([np.inf, -np.inf],np.nan)

#This line of code replaces infinite values (np.inf for positive infinity and -np.inf for negative infinity) in the 'ROI' 
#column of a DataFrame (df) with NaN (Not a Number).
```

### # Analyzing the Popularity of Movies by Month

### # Maping the month names to the respective indexes

```python
df5 = df.groupby('release_month')['popularity'].sum()
```

```python
df5.index
```

```python
df5.values
```

```python
data = {
    'release_month' : df5.index,
    'popularity':  df5.values
}
df5 = pd.DataFrame(data)
```

### # Analyzing correlation between different variables

```python
correlation = df['budget'].corr(df['revenue'])
print("Correlation = " , correlation)
print('A correlation of 0.729 means that movies with higher budgets tend to have higher profits, but there may still be exceptions or other factors (like marketing, quality, etc.) influencing the profit.')

# Scatter plot with a regression line between budget and revenue
plt.figure(figsize=(12,6))
sns.regplot(x='budget', y='revenue', data=df, scatter_kws={'s':5}, line_kws={"color": "blue","linewidth": 1})
plt.title('Budget vs Revenue Correlation')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
correlation = df['runtime'].corr(df['vote_average'])
print('Correlation =',correlation)
print('The correlation of 0.166 between runtime and vote average indicates a weak positive relationship, suggesting that longer movies have slightly higher ratings, but the effect is minimal.')
plt.figure(figsize=(12,6))
sns.regplot(x='runtime', y='vote_average', data=df, scatter_kws={'s':5}, line_kws={"color": "blue","linewidth": 1})
plt.title('Runtime vs Rating Correlation')
plt.xlabel("Runtime")
plt.ylabel('Rating')
plt.show()
```
